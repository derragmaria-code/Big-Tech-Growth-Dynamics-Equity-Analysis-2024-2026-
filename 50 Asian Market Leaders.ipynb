```python
import yfinance as yf
import pandas as pd
from scipy.stats import pearsonr
import time

# 1. The Asia-50 Universe (Tech, Semis, EV, and Industrials)
# Format: Ticker.Suffix (T=Tokyo, HK=Hong Kong, TW=Taiwan, KS=Korea)
tickers = [
    # TAIWAN (Semiconductors & Electronics)
    '2330.TW', '2317.TW', '2454.TW', '2308.TW', '2303.TW', '3711.TW', '2382.TW', '2412.TW',
    # JAPAN (Robotics, Automotive, Tech)
    '7203.T', '8035.T', '6758.T', '9984.T', '6857.T', '6501.T', '6702.T', '6981.T', '6954.T', '7741.T',
    '4063.T', '8001.T', '9432.T', '4502.T', '6367.T', '6503.T', '6723.T',
    # HONG KONG / CHINA (Internet & EV)
    '0700.HK', '9988.HK', '3690.HK', '1211.HK', '1810.HK', '9618.HK', '9888.HK', '2318.HK', '1299.HK',
    '0388.HK', '0941.HK', '1024.HK', '2015.HK', '2331.HK', '2269.HK',
    # SOUTH KOREA (Memory & Tech)
    '005930.KS', '000660.KS', '035420.KS', '035720.KS', '005380.KS', '005490.KS', '051910.KS', '373220.KS',
    '006400.KS', '000270.KS'
]

results = []

print(f"ðŸš€ Starting analysis on {len(tickers)} Asian Market Leaders...")

for symbol in tickers:
    try:
        t = yf.Ticker(symbol)
        # Fetching info and history
        info = t.info
        hist = t.history(period="1y")
        
        if hist.empty or 'revenueGrowth' not in info:
            continue
            
        # Metrics
        growth = info.get('revenueGrowth')
        start_price = hist['Close'].iloc[0]
        end_price = hist['Close'].iloc[-1]
        one_year_return = (end_price - start_price) / start_price
        
        # We only append if we have valid numerical data for both
        if growth is not None and one_year_return is not None:
            results.append({
                'Ticker': symbol,
                'Revenue_Growth': growth,
                'Return_1Y': one_year_return
            })
            
        # Respecting API limits with a tiny sleep
        time.sleep(0.1)
        
    except Exception as e:
        print(f"âš ï¸ Error fetching {symbol}: {e}")

# 2. Process Data
df = pd.DataFrame(results).dropna()

# 3. Statistical Calculation
if len(df) > 2:
    r_val, p_val = pearsonr(df['Revenue_Growth'], df['Return_1Y'])
    
    # 4. Final Output
    print("\n" + "="*50)
    print("ASIAN MARKET REGIME ANALYSIS (N=50)")
    print("="*50)
    print(df.sort_values(by='Return_1Y', ascending=False).to_string(index=False))
    print("-" * 50)
    print(f"Global Correlation (r): {r_val:.4f}")
    print(f"Statistical P-Value:    {p_val:.4f}")
    print("-" * 50)
    
    # 5. Skeptical Interpretation
    if p_val < 0.05:
        print("VERDICT: GEOGRAPHIC BIAS CONFIRMED.")
        print("Asia remains in a Growth-Dominant Regime while the US has decoupled.")
    else:
        print("VERDICT: GLOBAL REGIME SHIFT.")
        print("Revenue growth has failed as a predictive factor globally in 2026.")
else:
    print("Not enough data points collected.")
```
